```python
import pandas as pd
import os
```

```python
def list_csv_files(directory):
    csv_files = [file for file in os.listdir(directory) if file.endswith('.csv')]
    return csv_files

all_directory = 'All/'
all_csv_files = list_csv_files(all_directory)

baseline_directory = 'RMSE/'
baseline_csv_files = list_csv_files(baseline_directory)

def merge_csv_files(directory):
    csv_files = list_csv_files(directory)
    merged_df = pd.DataFrame()

    for file in csv_files:
        file_path = os.path.join(directory, file)
        df = pd.read_csv(file_path)
        df['file_name'] = file
        merged_df = pd.concat([merged_df, df], ignore_index=True)
    
    return merged_df



df_all_results = merge_csv_files(all_directory)
df_all_results = df_all_results.round(2)

df_baseline_results = merge_csv_files(baseline_directory)
df_baseline_results = df_baseline_results.round(2)
```

# Find differences Original VS ALL

```python
list_functions = df_all_results['Function'].unique().tolist()

df_diff = pd.DataFrame(columns=df_all_results.columns)

for current_function in list_functions:
    df_current_function = df_all_results[df_all_results['Function'] == current_function]
    df_current_function.reset_index(inplace=True, drop=True)
    df_current_result = df_current_function.copy()
    df_current_result[['A', 'B', 'C', 'D', 'E', 'A + B', 'RMSE', 'MSE', 'MAE', 'MAPE']] = df_current_function[['A', 'B', 'C', 'D', 'E', 'A + B', 'RMSE', 'MSE', 'MAE', 'MAPE']] - df_baseline_results[['A', 'B', 'C', 'D', 'E', 'A + B', 'RMSE', 'MSE', 'MAE', 'MAPE']]
    df_diff = pd.concat([df_diff, df_current_result], ignore_index=True)

df_diff
```

```python
# Función para colorear celdas según valor
def color_maximization_metric(val):
    if val > 0:
        color = 'background-color: lightgreen'
    elif val < 0:
        color = 'background-color: lightcoral'
    else:
        color = 'background-color: lightgrey'
    return color

def color_minimization_metric(val):
    if val < 0:
        color = 'background-color: lightgreen'
    elif val > 0:
        color = 'background-color: lightcoral'
    else:
        color = 'background-color: lightgrey'
    return color

def highlight_large_diff(threshold):
    def style(val):
        if abs(val) > threshold:
            return 'font-weight: bold'
        return ''
    return style

bold_style_func_1 = highlight_large_diff(1.0)

# Seleccionar solo las columnas _diff para aplicar estilos
maxi_columns = ['A + B']
mini_columns = ['RMSE', 'MSE', 'MAE', 'MAPE']

# Aplicar estilos
styled_diff = df_diff.style
styled_diff = styled_diff.applymap(color_maximization_metric, subset=maxi_columns)
styled_diff = styled_diff.applymap(color_minimization_metric, subset=mini_columns)
styled_diff = styled_diff.applymap(bold_style_func_1, subset=['A + B'])

# Detectar columnas numéricas
num_cols = (df_diff.select_dtypes(include='number').columns).values
```

```python
import numpy as np
num_cols = num_cols[num_cols != 'PH']  # Excluir PH
num_cols
```

```python
# Formatear con {:0.2f} → forzar 2 decimales incluso en ceros
styled_diff = styled_diff.format({col: "{:0.2f}" for col in num_cols})

styled_diff
```

```python
from openpyxl import load_workbook

# 1. Guardar el DataFrame normal sin formato especial
output_file = f"{allresult_name}_diff_{baseline_name}.xlsx"
styled_diff.to_excel(output_file, engine='openpyxl', index=False)

# 2. Abrir el archivo con openpyxl
wb = load_workbook(f"{allresult_name}_diff_{baseline_name}.xlsx")
ws = wb.active

# 3. Aplicar formato numérico a columnas específicas
from openpyxl.styles import numbers

# Definir columnas donde aplicar el formato
# num_cols = diff_df.select_dtypes(include='number').columns
col_indexes = [df_diff.columns.get_loc(col) + 1 for col in num_cols]  # openpyxl usa 1-based index

# Formatear cada celda de columnas numéricas
for row in ws.iter_rows(min_row=2, max_row=ws.max_row):  # saltar header
    for idx in col_indexes:
        cell = row[idx-1]
        cell.number_format = '0.00'  # formato real para Excel

# 4. Guardar
wb.save(output_file)
```
